```python
from collections import Counter
def solve(nums):
    count_map = Counter(nums)
    ans=0
    for key,val in count_map.items():
        if val>2:
            ans+=val//3
            val=val%3
        if val>1:
            ans+=val//2
            val=val%2
        if val ==1:
            return -1
    return ans
nums=[2,4,6,6,4,2,4]
print(solve(nums))
```

```python
from collections import Counter
def findPairs(lst, K): 
    res = []
    count = Counter(lst)
    for x in lst:
        y = K - x
        if (x != y and count[y]) or (x == y and count[y] > 1):
            res.append((x, y)) 
            count.subtract((x, y))      
    return res
lst = [1, 5, 3, 7, 9]
K = 12
print(findPairs(lst, K))
```

```python
#kendens algorithm
def maximumSubarraySum(arr):
    n = len(arr)
    maxSum = -999999999
    for i in range(0, n):
        currSum = 0
        for j in range(i, n):
            currSum = currSum + arr[j]
            if(currSum > maxSum):
                maxSum = currSum
    return maxSum
maximumSubarraySum(list(map(int,input().split())))
```

```python
def numDecodings(s: str) -> int:
    if len(s) == 0 or (len(s) == 1 and s[0] == '0'):
        return 0
    return numDecodingsHelper(s, len(s))
def numDecodingsHelper(s: str, n: int) -> int:
    if n == 0 or n == 1:
        return 1
    count = 0
    if s[n-1] > "0":
        count = numDecodingsHelper(s, n-1)
    if (s[n - 2] == '1'
        or (s[n - 2] == '2'
            and s[n - 1] < '7')):
        count += numDecodingsHelper(s, n - 2)
    return count
# Driver code
digits = "1101"
print("Count is ", numDecodings(digits))
```

```python
def numDecodings(s):
        if len(s) == 0 or s[0] == '0':
            return 0
        prev, prev_prev = 1, 0
        for i in range(len(s)):
            cur = 0
            if s[i] != '0':
                cur = prev
            if i > 0 and (s[i - 1] == '1' or (s[i - 1] == '2' and s[i] <= '6')):
                cur += prev_prev
            prev, prev_prev = cur, prev
        return prev
numDecodings("1101")
```

```python
def candy(ratings):
    count = 1
    candies = [1]
    for index in range(1, len(ratings)):
        if ratings[index] > ratings[index - 1]:
            count += 1
        else:
            count = 1
        candies.append(count)
    return sum(candies)
n=int(input())
ratings=list(map(int,input().split()))
candy(ratings)
```

```python
def removeDuplicates(s):
        stack=[]
        for i in s:
            if stack and stack[-1]==i:
                stack.pop()
            else:
                stack.append(i)
        return ''.join(stack)
removeDuplicates("abbca")
```
